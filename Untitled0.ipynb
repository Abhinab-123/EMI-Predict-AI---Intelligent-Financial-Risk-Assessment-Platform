## Handle missing values

### Subtask:
Identify and address missing values in the dataframe. This might involve imputation or removal based on the extent of missingness and the nature of the data.


**Reasoning**:
Identify columns with missing values and the number of missing values in each column.



```python
missing_values=df.isnull().sum()
print("missing values for columns")
print(missing_values[missing_values>0])
```

```python
for col in df.columns:
  if df[col].dtype=='float64' and missing_values[col]>0:
    print(f"{col} = float64 dtype")
```

**Reasoning**:
Determine an appropriate strategy to handle missing values for each column based on the data type and the percentage of missing values, and then apply the chosen strategy. Since the missing values are relatively few (less than 1%), imputing with the median for numerical columns and mode for categorical columns is a reasonable approach to preserve data and avoid removing a significant number of rows.



```python
# Convert columns to numeric, coercing errors
numerical_cols_to_impute = ['monthly_rent', 'credit_score', 'bank_balance', 'emergency_fund']
for col in numerical_cols_to_impute:
    df[col] = pd.to_numeric(df[col], errors='coerce')


# Impute numerical columns with the median
for col in numerical_cols_to_impute:
    df[col] = df[col].fillna(df[col].median())

# Impute categorical column with the mode
df['education'] = df['education'].fillna(df['education'].mode()[0])

# Verify that there are no remaining missing values
print("\nMissing values after imputation:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

## Create derived financial ratios

### Subtask:
Calculate debt-to-income, expense-to-income, and affordability ratios based on existing financial features.


**Reasoning**:
Calculate the debt-to-income, expense-to-income, and affordability ratios based on existing financial features, handling potential division by zero.



```python
# Calculate the total monthly expenses
df['total_monthly_expenses'] = df['monthly_rent'] + df['school_fees'] + df['college_fees'] + df['travel_expenses'] + df['groceries_utilities'] + df['other_monthly_expenses']

# Calculate the debt-to-income ratio, handling division by zero
df['debt_to_income_ratio'] = df['current_emi_amount'] / df['monthly_salary']
df['debt_to_income_ratio'] = df['debt_to_income_ratio'].replace([np.inf, -np.inf], np.nan).fillna(0) # Replace inf with NaN and then fill NaN with 0

# Calculate the expense-to-income ratio, handling division by zero
df['expense_to_income_ratio'] = df['total_monthly_expenses'] / df['monthly_salary']
df['expense_to_income_ratio'] = df['expense_to_income_ratio'].replace([np.inf, -np.inf], np.nan).fillna(0) # Replace inf with NaN and then fill NaN with 0

# Calculate the affordability ratio, handling division by zero
df['affordability_ratio'] = df['max_monthly_emi'] / df['monthly_salary']
df['affordability_ratio'] = df['affordability_ratio'].replace([np.inf, -np.inf], np.nan).fillna(0) # Replace inf with NaN and then fill NaN with 0

# Display the first few rows with the new columns
display(df[['monthly_salary', 'current_emi_amount', 'total_monthly_expenses', 'max_monthly_emi', 'debt_to_income_ratio', 'expense_to_income_ratio', 'affordability_ratio']].head())
```

## Generate risk scoring features

### Subtask:
Create features based on credit history and employment stability.


**Reasoning**:
Create new features for credit history risk and employment stability based on predefined ranges of 'credit_score' and 'years_of_employment' respectively, and then display the new columns.



```python
# Define credit score ranges and corresponding risk categories
credit_score_bins = [0, 580, 670, 850, np.inf]
credit_risk_labels = ['High', 'Medium', 'Low', 'Unknown'] # Assuming scores above 850 are less common or require special handling

# Create 'credit_history_risk' feature
df['credit_history_risk'] = pd.cut(df['credit_score'], bins=credit_score_bins, labels=credit_risk_labels, right=False)

# Define years of employment ranges and corresponding stability categories
employment_years_bins = [0, 2, 7, np.inf]
employment_stability_labels = ['Low', 'Medium', 'High']

# Create 'employment_stability' feature
df['employment_stability'] = pd.cut(df['years_of_employment'], bins=employment_years_bins, labels=employment_stability_labels, right=False)

# Display the first few rows with the new columns
display(df[['credit_score', 'credit_history_risk', 'years_of_employment', 'employment_stability']].head())
```
